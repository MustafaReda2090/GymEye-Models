```python
import cv2
import os

# Open the video file
video_path = "E:/FCI Bio/GP/Dataset/plank/videos/NEVER_DO_PLANKS_LIKE_THIS_10_Most_Common_Mistakes.mp4"
cap = cv2.VideoCapture(video_path)

# Create a directory to store the frames
frames_dir = "E:/FCI Bio/GP/Dataset/plank/final data"
os.makedirs(frames_dir, exist_ok=True)

# Loop through the video frames and save each one as an image file
frame_count = 0
while True:
    # Read the next frame from the video
    ret, frame = cap.read()
    if ret:
        # Save the frame as an image file
        frame_path = os.path.join(frames_dir, f"frame{frame_count:04d}.jpg")
        cv2.imwrite(frame_path, frame)
        frame_count += 1
    else:
        break

# Release the video file and print the number of frames extracted
cap.release()
print(f"Extracted {frame_count} frames to {frames_dir}")
```

```python
import cv2
import os

# Define the input and output directories
input_dir = "C:/Users/Alrowad/Exercise-correction/bicpes_model/dataset/test"
output_dir = "C:/Users/Alrowad/Exercise-correction/bicpes_model/dataset/testing"

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Loop through the video files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".mp4"):
        # Open the video file
        video_path = os.path.join(input_dir, filename)
        cap = cv2.VideoCapture(video_path)

        # Loop through the video frames and save each one as an image file
        frame_count = 0
        while True:
            # Read the next frame from the video
            ret, frame = cap.read()
            if ret:
                # Save the frame as an image file
                frame_path = os.path.join(output_dir, f"{filename}_frame{frame_count:04d}.jpg")
                cv2.imwrite(frame_path, frame)
                frame_count += 1
            else:
                break

        # Release the video file and print the number of frames extracted
        cap.release()
        print(f"Extracted {frame_count} frames from {filename}")

print("Done!")
```
